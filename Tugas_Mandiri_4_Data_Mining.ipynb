```python
import pandas as pd
df = pd.read_csv('/content/Online Sales Data.csv')
df.head()
```

```python
df['Product Name'].values
```

```python
data = list(df["Product Name"].apply(lambda x:x.split(",") ))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

data_encoded = te.fit(data).transform(data)

df = pd.DataFrame(data_encoded, columns=te.columns_)
df = df.astype(int)
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(df, min_support=0.001, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
rules
```
