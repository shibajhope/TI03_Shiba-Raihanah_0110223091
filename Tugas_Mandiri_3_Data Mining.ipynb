```python
import pandas as pd  # Importing pandas for DataFrame handling
import matplotlib.pyplot as plt  # Importing matplotlib for plotting

# Example DataFrame creation (replace this with your actual data loading)
data = {
    'Tinggi_Badan': [160, 158, 163, 167, 169, 162, 165, 170, 160, 162, 167]
}
df = pd.DataFrame(data)  # Creating the DataFrame

# Calculate the distribution of 'Tinggi_Badan'
Tinggi_Badan_distribution = df['Tinggi_Badan'].value_counts()

# Creating a pie chart
plt.figure(figsize=(8, 9))
plt.pie(Tinggi_Badan_distribution, labels=Tinggi_Badan_distribution.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue', 'lightgreen', 'lightyellow', 'lightpink'])
plt.title('Distribusi Tinggi Badan Idol')
plt.axis('equal')  # Ensure pie chart is a circle

# Display the pie chart
plt.show()
```
